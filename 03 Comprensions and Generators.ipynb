## generator function

```python
def return_num():
    for x in range(5):
        yield x
gen_num = return_num()
```

```python
next(gen_num)
```

```python
def counter():
    x=0
    while True :
        yield x
        x += 1
count = counter()
```

```python
next(count)
```

```python
def fib():
    for cur in (0,1):
        last = cur 
        yield cur
    while True:
        yield cur
        last,cur = cur,last+cur
```

```python
f = fib()
```

```python
next(f)
```
